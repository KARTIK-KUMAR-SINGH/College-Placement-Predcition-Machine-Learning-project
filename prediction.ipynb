```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("placement_dataset .csv")
df
```

```python
df.isnull().sum()
```

```python
df.drop(["Student_ID" , "CTC_LPA" , "Percentage_10th" , "Percentage_12th" , "Languages_Known"] , axis = 1 , inplace = True)
```

```python
df
```

```python
print(df['ML_AI_Skills'].unique())
print(df['Web_Dev_Skills'].unique())
print(df['Internship_Relevance'].unique())
print(df['Leadership_Roles'].unique())
```

```python
print(df['DSA_Proficiency'].unique())
print(df['Placed'].unique())
```

```python
df.fillna("None" , inplace=True)
```

```python
print(df['ML_AI_Skills'].unique())
print(df['Web_Dev_Skills'].unique())
print(df['Internship_Relevance'].unique())
print(df['Leadership_Roles'].unique())
```

```python
df
```

```python
transformer = ColumnTransformer(transformers =[
    ('tnf1', OneHotEncoder(drop = 'first' , sparse_output = False) , ['Branch'] ),
    ('tnf2' , OrdinalEncoder(categories = [['Beginner' , 'Intermediate' , 'Advanced']]) , ['DSA_Proficiency']),
    ('tnf3' , OrdinalEncoder(categories = [['None' , 'Basic' , 'Intermediate' , 'Advanced']]) , ['ML_AI_Skills']),
    ('tnf4' , OrdinalEncoder(categories = [['None' , 'Basic' , 'Intermediate' , 'Advanced']]) , ['Web_Dev_Skills']),
    ('tnf5' , OrdinalEncoder(categories = [['None' , 'Low' , 'Medium' , 'High']]) , ['Internship_Relevance']),
    ('tnf6' , OrdinalEncoder(categories = [['No' , 'Yes']]) , ['Leadership_Roles']),
],remainder = 'passthrough')
```

```python
transformer
```

```python
df
```

```python
x = df.iloc[ :  , : 19]
x
```

```python
le = LabelEncoder()
y = le.fit_transform(df['Placed'])
y
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2)
```

```python
y_train
```

```python
x_train_transformed = transformer.fit_transform(x_train)
x_test_transformed = transformer.transform(x_test)
```

```python
x_train_transformed
```

```python
x_train_transformed.shape
```

```python
print(type(x_train_transformed))
```

```python
x_train_df = pd.DataFrame(
    x_train_transformed ,
    columns = transformer.get_feature_names_out()
)
x_train_df
```

```python
model = DecisionTreeClassifier()
model.fit(x_train_transformed , y_train)
```

```python
y_pred = model.predict(x_test_transformed)
```

```python
print('Accuracy Score :-' , accuracy_score(y_pred , y_test))
```
